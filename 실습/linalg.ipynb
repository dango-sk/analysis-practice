```python
a = int(input())
b = int(input())

if a > 0 and b > 0:
    print(1)
elif a > 0 and b < 0:
    print(4)
elif a < 0 and b > 0:
    print(2)
elif a < 0 and b < 0:
    print(3)
```

```python
H, M = map(int, input().split())

if M < 45 :	# 분단위가 45분보다 작을 때 
    if H == 0 :	# 0 시이면
        H = 23
        M += 60
    else :	# 0시가 아니면 (0시보다 크면)
        H -= 1	
        M += 60
        
print(H, M-45)
```

```python
test =  list(map(int, input().split()))

if min(test) == max(test):
    print(10000 + (test[0]*1000))
elif (test[0] != test[1]) and (test[1] != test[2]) and (test[0] != test[2]):
    print(max(test) * 100)
else:
    print(1000+(sorted(test)[1]*100))
```
