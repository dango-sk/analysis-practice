```python
## Generating random numbers : inverse transform sampling

# 원하는 분포의 cdf F(x)의 역함수에 uniform 난수를 넣으면 
# 해당 분포의 난수가 튀어나옴 
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
#parameters
theta = 2.0 
n = 100000

#uniform random numbers
U = np.random.rand(n)

#inverse transform sampling
X = - (theta) * np.log(U)
```

```python
plt.hist(X, bins=50 , density=True, alpha=0.6, color='g', label = 'simulated' )

x_vals = np.linspace(0, 20, 100)
pdf = (1/theta) * np.exp(-x_vals/theta)
plt.plot(x_vals, pdf, 'r-', lw=2, label='theoretical')

plt.title('Exponential Distribution : Inverse Transform Sampling')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Acceptance-Rejection Sampling (beta distribution)
# - f(x) : target distribution (sampling 하고자 하는 분포)
# - g(x) : proposal distribution (쉽게 샘플링 가능한 분포)
# - C : f(x) <= C * g(x) 를 만족하는 상수
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, uniform
```

```python
a, b = 3,2 
f = lambda x: beta.pdf(x, a, b)  # target distribution
g = lambda x: uniform.pdf(x, 0, 1)  # proposal distribution
```

```python
x_vals = np.linspace(0,1,1000)
c = f(x_vals).max() # f(x)/g(x)의 최대값
```

```python
# Acceptance-Rejection Sampling 

n = 10000
accepted = []
while len(accepted) < n:
    x_proposal = np.random.uniform(0, 1) # x from uniform(0,1)
    u = np.random.uniform(0, 1) # u from uniform(0,1)
    if u < f(x_proposal)/(c * g(x_proposal)): # acceptance condition
        accepted.append(x_proposal)

accepted = np.array(accepted)
```

```python
plt.hist(accepted , bins = 50 , density=True, alpha=0.6, color='g', label = 'simulated' )
plt.plot(x_vals, f(x_vals), 'r-', lw=2, label='theoretical')
plt.title("Acceptance-Rejection Sampling : Beta Distribution")
plt.xlabel('x')
plt.ylabel('Density') 
plt.legend()
plt.show()
```
